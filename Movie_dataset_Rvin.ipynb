## Outliers

```python
df.boxplot()
```

```python
nun_outlier_df = df.copy()

revenue_Q1 = nun_outlier_df['revenue'].quantile(0.25)

revenue_Q3 = nun_outlier_df['revenue'].quantile(0.75)

revenue_IQR = revenue_Q3 - revenue_Q1

budget_Q1 = nun_outlier_df['budget'].quantile(0.25)

budget_Q3 = nun_outlier_df['budget'].quantile(0.75)

budget_IQR = budget_Q3 - budget_Q1

nun_outlier_df.drop(nun_outlier_df.loc[(nun_outlier_df['revenue'] < (revenue_Q1 - 1.5*revenue_IQR)) | (nun_outlier_df['revenue'] > (revenue_Q3 + 1.5*revenue_IQR))].index,
                    inplace= True)

nun_outlier_df.drop(nun_outlier_df.loc[(nun_outlier_df['budget'] < (budget_Q1 - 1.5*budget_IQR)) | (nun_outlier_df['budget'] > (budget_Q3 + 1.5*budget_IQR))].index,
                    inplace= True)
```

# Data cleaning

```python
data = nun_outlier_df.copy()
data['release_date'] = pd.to_datetime(data['release_date'])
```

```python
data.drop(data.loc[data['runtime'] == 0.].index, inplace = True)
data.drop(data.loc[data['production_companies'] == '[]'].index, inplace = True)
data.drop(data.loc[data['production_countries'] == '[]'].index, inplace = True)
data.drop(data.loc[data['spoken_languages'] == '[]'].index, inplace = True)
```

```python
for i in list(data.index):
  data.loc[i, 'production_companies'] = data.loc[i, 'production_companies'].replace('[', '').replace(']','')
  data.loc[i, 'production_countries'] = data.loc[i, 'production_countries'].replace('[', '').replace(']','')
  data.loc[i, 'spoken_languages'] = data.loc[i, 'spoken_languages'].replace('[', '').replace(']','')
  data.loc[i, 'crew'] = data.loc[i, 'crew'].replace('[', '').replace(']','')

data.head()
```

```python
for i in list(data.index):
  if 'u00' in data.loc[i, 'production_companies']:
    data.loc[i, 'production_companies'] = data.loc[i, 'production_companies'].encode().decode('unicode_escape')

  if 'u00' in data.loc[i, 'cast']:
    data.loc[i, 'cast'] = data.loc[i, 'cast'].encode().decode('unicode_escape')

  if 'u00' in data.loc[i, 'crew']:
    data.loc[i, 'crew'] = data.loc[i, 'crew'].encode().decode('unicode_escape')

  if 'u00' in data.loc[i, 'director']:
    data.loc[i, 'director'] = data.loc[i, 'director'].encode().decode('unicode_escape')

  if 'u00' in data.loc[i, 'spoken_languages']:
    data.loc[i, 'spoken_languages'] = data.loc[i, 'spoken_languages'].encode().decode('unicode_escape')
```

```python
# Extracting all the companies and countries involved in the data set

countries = ''
companies = ''

for i in list(data.index):
  for j in range(len(data.loc[i, 'production_companies'].split('"name": ')) - 1):
    companies += data.loc[i, 'production_companies'].split('"name": ')[j+1].split(
        ',')[0].replace('"', '')
    if j+1 != len(data.loc[i, 'production_companies'].split('"name": ')) - 1:
      companies += '_'

  data.loc[i, 'Companies'] = companies
  companies = ''

for i in list(data.index):
  for j in range(len(data.loc[i, 'production_countries'].split('"name": ')) - 1):
    countries += data.loc[i, 'production_countries'].split('"name": ')[j+1].split(
        ',')[0].replace('"', '').replace('}', '')
    if j+1 != len(data.loc[i, 'production_countries'].split('"name": ')) - 1:
      countries += '_'

  data.loc[i, 'Countries'] = countries
  countries = ''

# Extracting all the languages defined in the data set

languages = ''

for i in list(data.index):
  for j in range(len(data.loc[i, 'spoken_languages'].split('"iso_639_1": ')) - 1):
    languages += data.loc[i, 'spoken_languages'].split('"iso_639_1": ')[j+1].split(
        ',')[0].replace('"', '').replace(',', '')
    if j+1 != len(data.loc[i, 'spoken_languages'].split('"name": ')) - 1:
      languages += '_'

  data.loc[i, 'Languages'] = languages
  languages = ''
```

```python
all_countries = list(data['Countries'].str.split('_').explode().value_counts().index)
all_companies = list(data['Companies'].str.split('_').explode().value_counts().index)
all_directors = data['director'].unique()
```

**What actors make good movies?**

```python
data.loc[600]['cast']
```

```python
dataset = data.copy()
```

```python
first_last = []
name = ''

for i in list(dataset.index):
    cast = dataset.loc[i, 'cast'].split()
    for j in range(int(len(cast)/2)):
        first_last.append(cast[2*j] + ' ' + cast[2*j + 1])

    for j in first_last:
        name += j
        name += '_'

    dataset.loc[i, 'cast'] = name
    name = ''
    first_last.clear()
```

```python
all_actors = list(dataset['cast'].str.split('_').explode().value_counts().drop('').index)

Actors = pd.DataFrame(index = all_actors)
Actors.index = Actors.index.rename('Actors')

for i in all_actors:
    Actors.loc[i, 'Popularity'] = dataset.loc[data['cast'].str.contains(i)]['popularity'].sum()

Actors.sort_values(ascending = False, by = 'Popularity', inplace = True)
Actors.head()
```

```python
fig, ax = plt.subplots(figsize = (20, 15))

ax= sns.barplot(x = list(range(20)), y = Actors.iloc[:20, 0])
ax.set_xticks(ticks = list(range(20)), labels = list(Actors.index)[:20], rotation = 45);
ax2.grid()

ax.set_title('Popularity of actors')
plt.show()
```
